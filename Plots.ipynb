```r
library(ggplot2)
```

```r
Time<-c("13.979","19.477","24.364","30.599","39.298","90.443","144.4545","624.24")
```

```r
Num_file<-c(10,20,30,40,50,75,100,200)
```

```r
Time_trigram<-c(10.70,10.749,11.082,15.5015,16.894)
```

```r
Num_file_trigram<-c(1,2,3,4,5)
```

```r
Time_plot<-cbind(Num_file,Time)
Time_plot<-as.data.frame(Time_plot)
Time_plot
```

<h4>Inference:</h4>
<p>1.It is observed that as the number of files increase the time taken to calculate the bigrams also increase. Thus it is seen that SPARK takes more time as file count increases.</p>
<p>2.The line plot shows the time vs number of files.</p>


```r
plot(x = Num_file,y = Time,type="b" ,col="red",xlab = "Number of Files" ,ylab = "Time Taken in seconds")
```

```r
Time_plot_1<-cbind(Num_file_trigram,Time_trigram)
Time_plot_1<-as.data.frame(Time_plot_1)
Time_plot_1
```

```r
plot(x = Num_file_trigram,y = Time_trigram,type="b" ,col="blue",xlab = "Number of Files" ,ylab = "Time Taken in seconds")
```

<h4>Inference:</h4>
<p>1.It is observed that as the number of files increase the time taken to calculate the trigrams also increase. Thus it is seen that SPARK takes more time as file count increases.</p>
<p>2.The line plot shows the time vs number of files.</p>


<h4>CONCLUSION</h4>
<b>It can been seen from the plots and the data that trigram calculations are more time taking as compared to bigrams.Thus for higher number of files the trigram spends considerable amount of time with respect to bigrams</b>
